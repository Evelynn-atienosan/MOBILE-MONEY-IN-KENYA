```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
master = pd.read_csv(r"C:\Users\hp\EAST AFRICA FINANCIAL INCLUSION\financial_inclusion_africa.csv")
```

```python
master.head(5)
```

```python
master['Year'].unique()
```

```python
Top_10 = master.loc[master['Year']==2024,['Country','Account']].nlargest(10,'Account')

Top_10_MM = master.loc[master['Year']==2024,['Country','Mobile Money Account']].nlargest(10,'Mobile Money Account')

fig , ax = plt.subplots(2,1)

sns.set_theme(style = 'ticks')


sns.barplot( data = Top_10, x = 'Account', y = 'Country', ax =ax[0] ,hue = 'Account', palette = 'dark:b_r')
sns.despine()
ax[0].legend().remove()

ax[0].set_ylabel('')
ax[0].set_xlabel('Account Ownership')
ax[0].set_title('"Top African Countries: Account vs Mobile Money Ownership (2024)"') 
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

for bar in ax[0].patches:
    width = bar.get_width()
    ax[0].text(
        width + 0.5,
        bar.get_y() + bar.get_height() / 2,
        f'{width:.1f}%',
        va='center'
    )



sns.barplot( data = Top_10_MM, x ='Mobile Money Account', y = 'Country', ax =ax[1], hue = 'Mobile Money Account',palette = 'dark:b_r' ) 
sns.despine()
ax[1].legend().remove()
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylabel('')
ax[1].set_xlabel('Mobile Money Account Ownership')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

for bar in ax[1].patches:
    width = bar.get_width()
    ax[1].text(
        width + 0.5,
        bar.get_y() + bar.get_height() / 2,
        f'{width:.1f}%',
        va='center'
    )
ax[0].tick_params(axis='x', labelbottom=False)
ax[1].tick_params(axis='x', labelbottom=False)

fig.tight_layout()
```
