```python
master = df_final.copy()

datasets = [
   df_internet_users_long, 
   df_GDP_long, 
   df_mobile_subscription_long,
   df_bank_branches_long,
   df_poverty_ratio_long, 
   df_Urban_Population_long,df_final_bank_account
  
]

for dataset in datasets:
    master = master.merge(
        dataset,
        on=["Country","Code","Year"],
        how="left"
    )

master
```

```python
cols_to_num = [ "Account", 
               "Digitally enabled Account", 
               "Made a Digital Payment",
                 "Mobile Money Account", "Internet_users",
                   "GDP", "Mobile_cellular_subscription", 
                   "Commercial_bank_branches", "Poverty_ratio", 
                   "Urban_population", "Bank Account" ] 
master[cols_to_num] = master[cols_to_num].apply( lambda x: pd.to_numeric(x, errors="coerce").round(2) )
```

```python
master["Financial_Institution_Account_only"] = master["Account"] - (master["Mobile Money Account"])

master["Mobile_Money_Account_only"] = master["Account"] - (master["Bank Account"])

master["Both_Accounts"] = master["Account"] - (master["Financial_Institution_Account_only"] + master["Mobile_Money_Account_only"])
```
