```python
import matplotlib.pyplot as plt
import numpy as np

# Example columns expected in your data:
# Country, Year, Financial_Institution_Account_only,
# Financial institution and mobile money account,
# Mobile money account only

countries =master.loc[master['Year'] == 2024].nlargest(5, 'Account')['Country'].unique()

years = [2017,2021,2024]

fig, ax = plt.subplots(figsize=(12, 6))

bar_width = 0.22
group_gap = 0.35

colors = {
    'Financial_Institution_Account_only': '#20A7A8',
    'Both_Accounts': "#5B9BD5",
    'Mobile_Money_Account_only': '#8064A2'
}

x_positions = []
x_labels = []
country_centers = []

start = 0

for country in countries:
    country_data = master[master['Country'] == country]

    country_x = []

    for i, year in enumerate(years):
        row = country_data[country_data['Year'] == year]

        if row.empty:
            continue

        x = start + i * bar_width
        country_x.append(x)

        fi_only = row['Financial_Institution_Account_only'].values[0]
        both = row['Both_Accounts'].values[0]
        mm_only = row['Mobile_Money_Account_only'].values[0]

        total = fi_only + both + mm_only

        ax.bar(
            x,
            fi_only,
            width=bar_width,
            color=colors['Both_Accounts']
        )

        ax.bar(
            x,
            both,
            bottom=fi_only,
            width=bar_width,
            color=colors['Financial_Institution_Account_only']
        )

        ax.bar(
            x,
            mm_only,
            bottom=fi_only + both,
            width=bar_width,
            color=colors['Mobile_Money_Account_only']
        )

        ax.text(
            x,
            total + 2,
            f'{total:.0f}',
            ha='center',
            va='bottom',
            fontsize=10
        )

        x_positions.append(x)
        x_labels.append(str(year))

    country_centers.append(np.mean(country_x))
    start += len(years) * bar_width + group_gap

ax.set_xticks(x_positions)
ax.set_xticklabels(x_labels, rotation=90)

for center, country in zip(country_centers, countries):
    ax.text(
        center,
        -13,
        country,
        ha='center',
        va='top',
        fontsize=11
    )

ax.set_ylim(0, 105)
ax.set_ylabel('Adults with an account (%)')
ax.set_title('Account ownership growth in Sub-Saharan Africa', loc='left')

ax.legend(
    colors.keys(),
    loc='lower center',
    bbox_to_anchor=(0.5, -0.28),
    ncol=3,
    frameon=False
)

ax.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```
