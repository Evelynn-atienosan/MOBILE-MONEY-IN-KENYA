```python
import pandas as pd
import numpy as np

master = pd.read_csv(r"C:\Users\hp\EAST AFRICA FINANCIAL INCLUSION\financial_inclusion_africa.csv")
```

```python
print(master.columns)
```

```python

master["Population_adult"] = pd.to_numeric(master["Population_adult"], errors="coerce")

exclude_cols = ["Country", "Code", "Year", "Population_adult"]

indicator_cols = [
    col for col in master.columns
    if col not in exclude_cols
]

ssa_rows = []

for year, group in master.groupby("Year"):

    total_population = group["Population_adult"].sum(skipna=True)

    row = {
        "Region": "Sub-Saharan Africa",
        "Code": "SSA",
        "Year": year,
        "Population_adult": total_population
    }

    for col in indicator_cols:

        valid = group[[col, "Population_adult"]].copy()

        valid[col] = pd.to_numeric(valid[col], errors="coerce")
        valid["Population_adult"] = pd.to_numeric(valid["Population_adult"], errors="coerce")

        valid = valid.dropna(subset=[col, "Population_adult"])

        if valid.empty or total_population == 0:
            row[col] = pd.NA
            continue

        available_population = valid["Population_adult"].sum()
        missing_population = total_population - available_population

        if missing_population / total_population > (1 / 3):
            row[col] = pd.NA
        else:
            row[col] = (
                (valid[col] * valid["Population_adult"]).sum()
                / valid["Population_adult"].sum()
            )

    ssa_rows.append(row)

ssa_data = pd.DataFrame(ssa_rows)
```
