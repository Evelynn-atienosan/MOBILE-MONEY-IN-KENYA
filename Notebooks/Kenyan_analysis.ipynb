```python
import pandas as pd

# 1. Create the DataFrame using your exact column names
data = {
    "Year": [2021],
    "Saved any money  (% age 15+)": [66.7671320369555],
    "Saved at a bank or similar financial institution  (% age 15+)": [21.4531739032776],
    "Saved at a bank or similar financial institution or using a mobile money account  (% age 15+)": [45.0110991275483],
    "Saved money using a mobile money account  (% age 15+)": [37.2453264493391]
}

df_savers_2021 = pd.DataFrame(data)

# 2. Extract values dynamically from the DataFrame columns
saved_any = df_savers_2021["Saved any money  (% age 15+)"]
saved_bank = df_savers_2021["Saved at a bank or similar financial institution  (% age 15+)"]
saved_formal_or_mobile = df_savers_2021["Saved at a bank or similar financial institution or using a mobile money account  (% age 15+)"]
saved_mobile = df_savers_2021["Saved money using a mobile money account  (% age 15+)"]

# 3. Calculate underlying population overlaps (% of total population)
pop_both = saved_bank + saved_mobile - saved_formal_or_mobile
pop_mobile_only = saved_mobile - pop_both
pop_bank_only = saved_bank - pop_both
pop_informal_only = saved_any - saved_formal_or_mobile

# 4. Append the new calculations back as new columns scaled to the % of savers
df_savers_2021["Mobile Money ONLY (% of savers)"] = (pop_mobile_only / saved_any) * 100
df_savers_2021["Informal ONLY (% of savers)"] = (pop_informal_only / saved_any) * 100
df_savers_2021["Both Accounts (% of savers)"] = (pop_both / saved_any) * 100
df_savers_2021["Traditional Bank ONLY (% of savers)"] = (pop_bank_only / saved_any) * 100

df_savers_2021

```

```python
import pandas as pd

# 1. Store your 2024 row exactly as it is structured in the Global Findex
private_wages_2024 = {
    "Year": [2024],
    "Received private sector wages  (% age 15+)": [26.2883447955206],
    "Received private sector wages: into an account (% of private sector wage recipients, age 15+)": [81.1812050561477],
    "Received private sector wages: through a mobile phone (% of private sector wage recipients, age 15+)": [70.7290405709424]
}

df_private_wages_2024 = pd.DataFrame(private_wages_2024)

# 2. Extract active columns dynamically using unique wage variable tags
wages_any_2024 = df_private_wages_2024["Received private sector wages  (% age 15+)"]
into_account_rate_2024 = df_private_wages_2024["Received private sector wages: into an account (% of private sector wage recipients, age 15+)"]
through_mobile_rate_2024 = df_private_wages_2024["Received private sector wages: through a mobile phone (% of private sector wage recipients, age 15+)"]

# 3. Step 1: Scale the recipient ratios back into total population percentages
pop_into_account_2024 = (into_account_rate_2024 / 100.0) * wages_any_2024
pop_mobile_2024 = (through_mobile_rate_2024 / 100.0) * wages_any_2024

# 4. Step 2: Run subtractions against the master benchmarks to fill in the missing '..' data
# Since "into an account" includes ALL electronic methods, the remainder is Cash Only
pop_cash_only_2024 = wages_any_2024 - pop_into_account_2024

# Subtract mobile phone wages from total account wages to isolate pure Bank/Card options
pop_bank_or_card_only_2024 = pop_into_account_2024 - pop_mobile_2024

# 5. Save calculations back to the DataFrame using clean, shared column names
df_private_wages_2024["Cash ONLY"] = pop_cash_only_2024
df_private_wages_2024["Mobile Money / Phone"] = pop_mobile_2024
df_private_wages_2024["Traditional Bank or Card ONLY"] = pop_bank_or_card_only_2024


df_private_wages_2024
```
